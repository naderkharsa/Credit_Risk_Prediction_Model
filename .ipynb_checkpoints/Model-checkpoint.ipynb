## Content
#### There are 25 variables:

-ID: ID of each client

-LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit

-SEX: Gender (1=male, 2=female)

-EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)

-MARRIAGE: Marital status (1=married, 2=single, 3=others)

-AGE: Age in years

-PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, … 
8=payment delay for eight months, 9=payment delay for nine months and above)

-PAY_2: Repayment status in August, 2005 (scale same as above)

-PAY_3: Repayment status in July, 2005 (scale same as above)

-PAY_4: Repayment status in June, 2005 (scale same as above)

-PAY_5: Repayment status in May, 2005 (scale same as above)

-PAY_6: Repayment status in April, 2005 (scale same as above)

-BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)

-BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)

-BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)

-BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)

-BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)

-BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)

-PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)

-PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)

-PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)

-PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)

-PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)

-PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)
##### TARGET : Default.payment.next.month: Default payment (1=yes, 0=no)

```python
pay_columns = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
df[pay_columns] = df[pay_columns].replace(-1, 0) # replace -1 with 0 //// 0 =' pay duly'
```

```python
df['SUM_PAY'] = df[pay_columns].sum(axis=1)
df['MAX_PAY'] = df[pay_columns].max(axis=1)
df['MEAN_PAY'] = df[pay_columns].mean(axis=1)
df['DELAY_FREQUENCY'] = (df[pay_columns] > 0).sum(axis=1)
```

```python
df['COMBINED_PAY'] = df[pay_columns].astype(str).agg('-'.join, axis=1)

# Create dummy variables
#dummies = pd.get_dummies(df['COMBINED_PAY'], prefix='PAY')
#df = pd.concat([df, dummies], axis=1)
```

```python
df.drop('COMBINED_PAY', axis=1, inplace=True)
```

```python
bill_columns = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
```

```python
df['TOTAL_BILL'] = df[bill_columns].sum(axis=1)
df['AVG_BILL'] = df[bill_columns].mean(axis=1)
df['STD_BILL'] = df[bill_columns].std(axis=1)
df['MAX_BILL'] = df[bill_columns].max(axis=1)
df['MIN_BILL'] = df[bill_columns].min(axis=1)
```

```python
# Calculate the trend (linear regression slope) of the bill amounts
def calculate_trend(row):
    x = np.arange(1, 7)
    y = row[bill_columns]
    slope = np.polyfit(x, y, 1)[0]
    return slope

df['BILL_TREND'] = df.apply(calculate_trend, axis=1)
df['BILL_TREND']
```

```python
pay_amt_columns = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
```

```python
df['TOTAL_PAY_AMT'] = df[pay_amt_columns].sum(axis=1)
df['AVG_PAY_AMT'] = df[pay_amt_columns].mean(axis=1)
df['STD_PAY_AMT'] = df[pay_amt_columns].std(axis=1)
df['MAX_PAY_AMT'] = df[pay_amt_columns].max(axis=1)
df['MIN_PAY_AMT'] = df[pay_amt_columns].min(axis=1)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['SEX', 'EDUCATION', 'MARRIAGE'], drop_first=True)
```

```python
columns_to_convert =['SEX_1', 'EDUCATION_2', 'EDUCATION_3', 'EDUCATION_4','MARRIAGE_2']
```

```python
for col in columns_to_convert: 
    df[col] = df[col].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_columns = ['LIMIT_BAL', 'AGE', 'TOTAL_BILL', 'AVG_BILL', 'STD_BILL', 'MAX_BILL', 'MIN_BILL', 
               'TOTAL_PAY_AMT', 'AVG_PAY_AMT', 'STD_PAY_AMT', 'MAX_PAY_AMT', 'MIN_PAY_AMT']
df[num_columns] = scaler.fit_transform(df[num_columns])
```

## Key Observations:
### Positive Correlations with Default Payment:

#### -PAY_0, PAY_2, PAY_3, PAY_4, PAY_5, PAY_6: These features (representing the repayment status) show positive correlations, indicating that delays in payments are strong predictors of default.
#### -DELAY_FREQUENCY: High positive correlation, reinforcing that frequent delays are a strong indicator of default.
#### -SUM_PAY, MAX_PAY, MEAN_PAY: These aggregated delay metrics are also positively correlated with default payment.

### Negative Correlations with Default Payment:

#### -LIMIT_BAL: Negative correlation, suggesting that higher credit limits might be associated with lower default risk.
#### -TOTAL_BILL, AVG_BILL, MAX_BILL: Negative correlations indicate that higher bill amounts might be linked to lower default probabilities, potentially reflecting better financial capacity.

# Building a prediction model 

```python
df.columns
```

```python
X = df.drop(columns=['ID', 'default.payment.next.month'])
y = df['default.payment.next.month']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## using LogisticRegression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
# Evaluate the model
print("Logistic Regression Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
print(confusion_matrix(y_test, pred))
```

## using RandomForestClassifier 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
pred2 = rf.predict(X_test)
```

```python
# Evaluate the model
print("Random Forest Accuracy:", accuracy_score(y_test, pred2))
print(classification_report(y_test, pred2))
print(confusion_matrix(y_test, pred2))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [10, 20, 30, None],
    'criterion': ['gini', 'entropy']
}
```

```python
grid_rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1)
grid_rf.fit(X_train, y_train)
```

```python
# Best parameters
print("Best parameters found: ", grid_rf.best_params_)
```

```python
best_params = grid_rf.best_params_
```

```python
# Train the model with the best parameters
rf_tuned = RandomForestClassifier(n_estimators=best_params['n_estimators'],
                                  criterion=best_params['criterion'],
                                  max_depth=best_params['max_depth'],
                                  max_features=best_params['max_features'],
                                  random_state=42)
```

```python
rf_tuned.fit(X_train, y_train)
```

```python
# Predictions
y_pred_tuned = rf_tuned.predict(X_test)
```

```python
# Evaluate the model
print("Tuned Random Forest Accuracy:", accuracy_score(y_test, y_pred_tuned))
print(classification_report(y_test, y_pred_tuned))
print(confusion_matrix(y_test, y_pred_tuned))
```
